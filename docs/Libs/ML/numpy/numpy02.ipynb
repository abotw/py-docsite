## Numpy Array Attributes

```python
import numpy as np

# seed for reproducibility
np.random.seed(0) 

# one-dimensional array
x1 = np.random.randint(10, size=6)
# two-dimensional array
x2 = np.random.randint(10, size=(3, 4))
# three-dimensional array
x3 = np.random.randint(10, size=(3, 4, 5))

print("x3.ndim:\t", x3.ndim)
print("x3.shape:\t", x3.shape)
print("x3.size:\t", x3.size)
print("x3.dtype:\t", x3.dtype)
print("x3.itemsize:\t", x3.itemsize, "bytes")
print("x3.nbytes:\t", x3.nbytes, "bytes")
```

## Array Slicing: Accessing Subarrays

### One-dimensional subarrays

```python
x= np.arange(10)
x
```

```python
# first five elements
x[:5]
```

```python
# elements after index 5
x[5:]
```

```python
# middle sub-array
x[4:7]
```

```python
# every other element
x[::2]
```

```python
# every other element, starting at index 1
x[1::2]
```

```python
# all elements, reversed
x[::-1]
```

```python
# reserved every other from index 5
x[5::-2]
```

### Multi-dimensional subarrays

```python
x2
```

```python
# two rows, three columns
x2[:2, :3]
```

```python
# all rows, every other column
x2[:3, ::2]
```

```python
x2[::-1, ::-1]
```

#### Accessing array rows and columns

```python
# first column of x2
print(x2[:, 0])
```

```python
# first row of x2
print(x2[0, :])
```

```python
# equivalent to x2[0, :]
print(x2[0])
```

### Subarrays as no-copy views

```python
print(x2)
```

```python
x2_sub = x2[:2, :2]
print(x2_sub)
```

```python
x2_sub[0, 0] = 99
print(x2_sub)
```

```python
print(x2)
```

### Creating copies of arrays

```python
x2_sub_copy = x2[:2, :2].copy()
print(x2_sub_copy)
```

```python
x2_sub_copy[0, 0] = 42
print(x2_sub_copy)
```

```python
print(x2)
```

## Reshaping of Arrays

```python
grid = np.arange(1, 10).reshape((3, 3))
print(grid)
```

```python
x = np.array([1, 2, 3])

# row vector via reshape
x.reshape((1, 3))
```

```python
# row vector via newaxis
x[np.newaxis, :]
```

```python
# column vector via reshape
x.reshape((3, 1))
```

```python
# column vector via newaxis
x[:, np.newaxis]
```

## Array Concatenation and Splitting

### Concatenation of Arrays

```python
x = np.array([1, 2, 3])
y = np.array([3, 2, 1])
np.concatenate([x, y])
```

```python
z = [99, 99, 99]
print(np.concatenate([x, y, z]))
```

```python
grid = np.array([[1, 2, 3],
                 [4, 5, 6]])
```

```python
# concatenate along the first axis
np.concatenate([grid, grid])
```

```python
# concatenate along the second axis (zero-indexed)
np.concatenate([grid, grid], axis=1)
```

---

```python
x = np.array([1, 2, 3])
grid = np.array([[9, 8, 7],
                 [6, 5, 4]])

# vertically stack the array
np.vstack([x, grid])
```

```python
# horizontally stack the arrays
y = np.array([[99],
              [99]])
np.hstack([grid, y])
```

### Splitting of arrays

```python
x = [1, 2, 3, 99, 99, 3, 2, 1]
x1, x2, x3 = np.split(x, [3, 5])
print(x1, x2, x3)
```

```python
grid = np.arange(16).reshape((4, 4))
grid
```

```python
upper, lower = np.vsplit(grid, [2])
print(upper)
print(lower)
```

```python
left, right = np.hsplit(grid, [2])
print(left)
print(right)
```
