## imshow

```python
import matplotlib.pyplot as plt
import numpy as np

n = 4

# create an nxn numpy array
a = np.reshape(np.linspace(0, 1, n**2), (n, n))

plt.figure(figsize=(12, 4.5))

# use imshow to plot the array
plt.subplot(131)
plt.imshow(a, # numpy array generating the image
           cmap='gray', # color map used to specify colors
           interpolation='nearest' # algorithms used to blend square colors, with 'nearest' colors will not be blended
          )
plt.xticks(range(n))
plt.yticks(range(n))
plt.title('gray color map, no blending', y=1.02, fontsize=12)

# the same array as above, but with different color map
plt.subplot(132)
plt.imshow(a,
           cmap='viridis',
           interpolation='nearest'
          )
plt.xticks(range(n))
plt.yticks()
plt.title('viridis color map, no blending', y=1.02, fontsize=12)

# the same array as above, but with blending
plt.subplot(133)
plt.imshow(a,
           cmap='viridis',
           interpolation='bicubic'
          )
plt.xticks(range(n))
plt.yticks()
plt.title('viridis color map, bicubic blending', y=1.02, fontsize=12)

plt.show()
```

## imsave

```python
n = 500

# create an nxn numpy array
a = np.random.random((n, n))

# save the image to a file 'test.png' using the 'jet' color map to specify colors
plt.imsave('test.png', a, cmap='inferno')
```

## imread

```python
img_array = plt.imread('tiger.webp')
print(img_array)
```

```python
img = img_array/255
plt.figure(figsize=(10, 10))
plt.imshow(img)
plt.show()
```

```python
plt.figure(figsize=(10,6))

for i in range(1,5):
    plt.subplot(2,2,i)
    x = 1 - 0.2*(i-1)
    plt.axis('off') # hide coordinate axes
    plt.title('x={:.1f}'.format(x))
    plt.imshow(img*x)

plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.imshow(img[:300, 500:800, :])
plt.axis('off')
plt.show()
```

```python
red_img = img.copy()

red_img[:, :, [1,2]] = 0

plt.figure(figsize=(10, 10))
plt.imshow(red_img)
plt.axis('off')
plt.show()
```
