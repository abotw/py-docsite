## Code Example

```python
import time

# 获取时间戳
print(time.time())

# 获取当前时间（时间元组）
print(time.localtime())

# 时间戳 -> 时间元组
print(time.localtime(time.time()))

# 获取格式化的时间
print(time.asctime(time.localtime()))

# 时间元组 -> 时间戳
print(time.mktime(time.localtime()))
```
