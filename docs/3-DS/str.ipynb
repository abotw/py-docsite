# String

## split

```python
str = "[哥伦比亚] 加西亚·马尔克斯 / 范晔 / 南海出版公司 / 2011-6 / 39.50元"
splited_lst = [item.strip() for item in str.split('/')]
print(splited_lst)
```

## replace

```python
replaced = str.replace('[', '【').replace(']', '】')
print(replaced)
```

## join

```python
authors_lst = splited_lst[-4::-1]

# authors_rlst = authors_lst[::-1]
authors_rlst = list(reversed(authors_lst))

authors_str = " | ".join(authors_rlst)
print(authors_str)
```

## format

```python
str_to_format = "{}-xxx"
for i in range(10):
    str = str_to_format.format(i)
    print(str)
```
