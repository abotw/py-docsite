## for-loop

```python
# for-loop over list
words = ['thus', 'solving', 'the', 'problem']
for word in words:
    print(word)
```

```python
lst = ['a', 'b', 'c', 'd']
lst.pop() # default = remove from end
```

```python
lst = ['a', 'b', 'c', 'd']
```

## extend()

```python
lst = [1, 2, 3]
x = [4, 5]
lst.extend(x) # extend = add all
lst
```

## +

```python
lst = [1, 2, 3]
x = [4, 5]
lst + x # make new combined list
```

```python
lst
```

```python
lst += x # += modifiles and adds
lst
```
