```python
nums = [1, 2, 3, 4, 5, 6]
[n*n for n in nums]
```

```python
[2**n for n in nums]
```

```python
[n*-1 for n in nums]
```

```python
[str(n)+'!' for n in nums]
```

```python
[n*n for n in nums if n>=3]
```
