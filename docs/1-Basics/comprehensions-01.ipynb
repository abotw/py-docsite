# Comprehensions-01

## 1. What Is a Comprehension?

A **comprehension** is a **short and readable way** to create a new collection from an existing one.

Instead of writing:

* many lines
* with loops
* and temporary variables

You write **one clear expression**.

Think of comprehension as:

> ðŸ” *â€œfor-loop + condition + resultâ€ in one line.*

## 2. Why Use Comprehensions?

Compared to normal loops, comprehensions are:

* Shorter
* More readable (once you get used to them)
* Often faster
* Very â€œPythonicâ€

## 3. List Comprehension (Most Important)

### Basic Syntax

```python
[new_value for item in iterable]
```

### Example 1: Squares of Numbers

**Normal way:**

````python
result = []
for x in range(5):
    result.append(x * x)
````

**List comprehension:**

````python
result = [x * x for x in range(5)]
````

âœ” Same result

âœ” Much shorter

### Example 2: With Condition

````python
evens = [x for x in range(10) if x % 2 == 0]
````

Equivalent loop:

````python
evens = []
for x in range(10):
    if x % 2 == 0:
        evens.append(x)
````

## 4. Understanding the Order (Very Important)

Comprehension order:

```text
result = [expression for item in iterable if condition]
```

Read it as:

> â€œFor each item in iterable, if condition is true, put expression into result.â€

## 5. Using `if ... else` in Comprehensions

### Example: Label Numbers

````python
labels = ["even" if x % 2 == 0 else "odd" for x in range(5)]
labels
````

âš ï¸ Note:

* `if/else` comes **before** `for`
* This is different from filtering

## 6. Nested List Comprehensions

### Example: Flatten a List

````python
matrix = [[1, 2], [3, 4], [5, 6]]
flat = [x for row in matrix for x in row]
flat
````

Equivalent loop:

````python
flat = []
for row in matrix:
    for x in row:
        flat.append(x)
flat
````

Read left â†’ right like nested loops.

## 7. Dictionary Comprehension

### Basic Syntax

```python
{key: value for item in iterable}
```

---

### Example: Squares Dictionary

```python
squares = {x: x * x for x in range(5)}
```

---

### With Condition

```python
even_squares = {x: x * x for x in range(10) if x % 2 == 0}
```

## 8. Set Comprehension

### Syntax

```python
{expression for item in iterable}
```

---

### Example

```python
unique_lengths = {len(word) for word in ["apple", "banana", "pear"]}
```

Result:

```python
{4, 5, 6}
```

## 9. Generator Expressions (Memory Friendly)

### Syntax

```python
(expression for item in iterable)
```

---

### Example

```python
gen = (x * x for x in range(1000000))
```

* Does NOT create a list
* Generates values one by one
* Saves memory

Use with:

```python
sum(gen)
```

## 10. When NOT to Use Comprehensions

Avoid comprehensions when:

* Logic is too complex
* Multiple nested conditions
* Side effects (printing, writing files)

âŒ Bad example:

```python

```

âœ” Better:

```python

for x in data:
    if x > 0 and x < 100 and x % 3 == 0:
        result.append(x)
```
