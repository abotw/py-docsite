## range(n)

```python
s = 'Python'
len(s)
```

```python
for i in range(len(s)):
    print(i, s[i])
```

## reversed()

```python
s = 'Python'
for i in reversed(range(len(s))):
    print(i, s[i])
```

## range(start, stop)

```python
list(range(5, 10))
```

```python
list(range(5, 7))
```

```python
list(range(5, 6))
```

```python
# start >= stop, no numbers
list(range(5, 5))
```

```python
list(range(0, 5))
```

```python
list(range(0, -1))
```

## range(start, stop, step))

```python
list(range(0, 10, 2))
```

```python
list(range(0, 10, 6))
```

```python
list(range(200, 800, 100))
```

```python
list(range(10, 5, -1))
```

```python
list(range(10, 5, -2))
```

```python
list(range(6, 5, -2))
```

```python
list(range(5, 5, -2))  # equal to stop is omitted
```

```python
list(range(4, 5, -2))  # beyond the stop is omitted
```
