# Enumerate

- <https://www.uvm.edu/~cbcafier/cs1210/book/11_loops/enumerate.html>

```python
incr = 5
lst = [1, 2, 3, 4]

for i, e in enumerate(lst):
    lst[i] = e + incr
lst
```

```python
lst = ['a', 'b', 'c', 'd']
for i, elem in enumerate(lst):
    print(f"The element at index {i} is '{elem}'.")
```
